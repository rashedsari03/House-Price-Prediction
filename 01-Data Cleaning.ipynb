# Data Cleaning

### Removing columns with null values > half the length of dataframe

```python
# Removing columns with null values > half the length of dataframe
columns_to_drop = [column for column in df_data.columns if df_data[column].isnull().sum() > df_data.shape[0] / 2]
columns_to_drop
```

```python
df_data.drop(columns_to_drop, axis=1, inplace=True)
df_data.info()
```

### Cleaning ('price_euro') column

```python
df_data.drop('price', axis=1, inplace=True)

# Drop rows with no price
df_data = df_data[df_data['price_euro'] != 'Price on application']

# Cleaning ('price_euro') column
df_data['price_euro'] = df_data['price_euro'].apply(lambda x: x.replace('€ ', '').replace('from ', '').replace(',', '').replace('.00', ''))

# Filter out rows where 'price_euro' column contains non-numeric values
df_data = df_data[df_data['price_euro'].str.isnumeric() != False]

# Convert the 'price_euro' column to float data type
df_data['price_euro'] = df_data['price_euro'].astype('float')


df_data['price_euro']
```

```python
# Cleaning ('number_rooms') column
df_data['number_rooms'].value_counts() # Values are corrupted
```

### Cleaning ('number_rooms') column

```python
# Extract the number of rooms from the 'rooms' column, ignoring any missing values.
# Split the values based on whitespace and retrieve the first part, which represents the number of rooms.
df_data['number_rooms'] = df_data['rooms'].dropna().apply(lambda x : x.split()[0])

# Remove the '+' sign from values representing '5+' rooms, considering them as 5 rooms.
# This standardizes the representation of rooms for consistency.
df_data['number_rooms'] = df_data['number_rooms'].str.replace('+', '')

# Drop rows with missing values in the 'number_rooms' column to ensure data integrity.
df_data = df_data[~df_data['number_rooms'].isnull()]

# Convert the values in the 'number_rooms' column to integer data type for numerical operations.
df_data['number_rooms'] = df_data['number_rooms'].astype('int')

df_data['number_rooms']
```

### Cleaning ('area_m2') columns


```python
# Extract the surface area from the 'surface' column, ignoring any missing values.
df_data['area_m2'] = df_data['surface'].dropna().apply(lambda x: x.split()[0])

# Remove any commas from the values to ensure consistency.
df_data['area_m2'] = df_data['area_m2'].str.replace(',', '')

# Drop the 'surface' column from the DataFrame since we have extracted the surface area.
df_data.drop('surface', axis=1, inplace=True)

# Convert the values in the 'area_m2' column to integer data type for numerical operations.
df_data['area_m2'] = df_data['area_m2'].astype('int')

# Display the 'area_m2' column to check the results.
df_data['area_m2']
```

### Cleaning ('bathrooms') column

```python
# Cleaning ('bathrooms') column
df_data['bathrooms'].value_counts() # Values are corrupted
```

```python
# Extracting the right number of bathrooms from the 'rooms' column

# For apartments with no bathroom information, assume they have 1 bathroom
# Extract the number of bathrooms using split and consider the last two characters of the split result.
# If ' bathroom' is in the value, extract the number before it, otherwise assume 1 bathroom.
df_data['bathrooms'] = df_data['rooms'].dropna().apply(lambda x: x.split(' bathroom')[0][-2:] if ' bathroom' in x else '1')

# Consider apartments with 3+ bathrooms as having just 3 bathrooms
df_data['bathrooms'] = df_data['bathrooms'].str.replace('+', '')

# Drop the 'rooms' column since we have extracted the necessary information into the 'bathrooms' column
df_data.drop('rooms', axis=1, inplace=True)

# Convert the values in the 'bathrooms' column to integer data type for numerical operations
df_data['bathrooms'] = df_data['bathrooms'].astype('int')

# Display the 'bathrooms' column to check the results
df_data['bathrooms']
```

```python
# Display value counts to understand the distribution of the number of bathrooms
df_data['bathrooms'].value_counts()
```

### Extracting useful information from ('floor') column

```python
# Create a new column 'with_disabled_access' to indicate if the property has disabled access
# Assign 1 if the string 'with disabled access' is present in the 'floor' column, otherwise 0.
df_data['with_disabled_access'] = df_data['floor'].dropna().apply(lambda x: 1 if 'with disabled access' in x else 0)

# Create a new column 'with_lift' to indicate if the property has a lift
# Assign 1 if the string 'with lift' is present in the 'floor' column, otherwise 0.
df_data['with_lift'] = df_data['floor'].dropna().apply(lambda x: 1 if 'with lift' in x else 0)

# Drop ('floor') column
df_data.drop('floor', axis=1, inplace=True)

df_data['with_disabled_access'].isna().sum()
# Remove null values
df_data = df_data[~df_data['with_disabled_access'].isnull()]
```

### Cleaning ('floor_number') column


```python
df_data['floor_number'].value_counts()
```

```python
# Extracting floor numbers from the 'floor_number' column and standardizing representations

#'M' is replaced with '0.5' (representing a mezzanine floor), 'SB' and 'B' are both replaced with '-1' (representing basement floors), and 'G' is replaced with '0' (representing ground floors).
# Replace 'M' with '0.5', 'SB' with '-1', 'B' with '-1', and 'G' with '0'
df_data['floor_number'] = df_data['floor_number'].dropna().apply(lambda x: x.replace('M', '0.5').replace('SB', '-1').replace('B', '-1').replace('G', '0'))
df_data['floor_number'] = df_data['floor_number'].dropna().apply(lambda x: x.split()[0])
# Convert the values in the 'floor_number' column to a categorical data type
df_data['floor_number'] = df_data['floor_number'].astype('category')

# Remove rows with null values in the 'floor_number' and 'description' columns
df_data = df_data[~df_data['floor_number'].isnull()]
df_data = df_data[~df_data['description'].isnull()]
```

```python
# Display the value counts of the 'floor_number' column to check the distribution
df_data['floor_number'].value_counts()
```

```python
# Make new column for the date of the announcement
df_data['listing Date'] = df_data['Reference and listing Date'].dropna().apply(lambda x : x.split('-')[-1])
df_data['listing Date'] = pd.to_datetime(df_data['listing Date'])
```

### Cleaning ('total building floors') column

```python
df_data['total building floors'].value_counts()
```

```python
df_data['total building floors'] = df_data['total building floors'].dropna().apply(lambda x : x.split(' ')[0])

# convert values to float
df_data['total building floors'] = df_data['total building floors'].dropna().astype('float')
df_data['total building floors']
```

### Cleaning ('condominium fees') column


```python
df_data['condominium fees'].value_counts()
```

```python
df_data['condominium fees'] = df_data['condominium fees'].dropna().apply(lambda x : x.replace('No condominium fees', '0').replace('/month', '').replace('€ ', '').replace(',', ''))

# Convert values to float
df_data['condominium fees'] = df_data['condominium fees'].astype('float')
```

```python
# ('condominium fees') column has alot of null values
df_data['condominium fees'].isnull().sum()
```

```python
# Getting the mean of the condominium fees of each neighborhood
means = df_data.groupby('neighborhood')['condominium fees'].mean()
means
```

```python
# Filling the null values with the mean of each neighborhood
new_values = means[df_data[df_data['condominium fees'].isnull()]['neighborhood']].values
new_values
```

```python
condition  = df_data['condominium fees'].isnull()
subset = df_data[condition]
subset
```

```python
df_data.loc[condition, 'condominium fees'] = new_values
```

```python
# Resetting the index of the DataFrame to a sequential range starting from zero.
# This ensures that the index of the DataFrame is renumbered to be consecutive integers starting from zero,
# reflecting the new order of rows after any operations that might have changed the original order.

df_data.index = range(df_data.shape[0])
```

```python
# Saving the cleaned DataFrame to a CSV file named 'house_prices_raw_data_cleaned.csv'.
# This allows for easy storage and sharing of the cleaned data for further analysis or use in other applications.
df_data.to_csv('house_prices_data_cleaned.csv')
```
